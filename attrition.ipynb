## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data

selected_columns = [
    'Age', 'BusinessTravel', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction',
    'HourlyRate', 'JobLevel', 'JobSatisfaction', 'JobRole', 'MaritalStatus'
]



# Create X_df using your selected columns
X_df = attrition_df[selected_columns]
print(X_df.head())
# Show the data types for X_df
X_df.dtypes
print(X_df.dtypes)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in ['BusinessTravel', 'JobRole', 'MaritalStatus']:
    X_train[column] = label_encoder.fit_transform(X_train[column])
    X_test[column] = label_encoder.transform(X_test[column])
# Verify the changes
X_train.head()
X_test.head()
```

```python
# Create a StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled[:5])
print(X_test_scaled[:5])
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_dept = OneHotEncoder(sparse=False)
# Fit the encoder to the training data
onehot_encoder_dept.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dept_encoded = onehot_encoder_dept.transform(y_train[['Department']])
y_test_dept_encoded = onehot_encoder_dept.transform(y_test[['Department']])
print(y_train_dept_encoded[:5])
print(y_test_dept_encoded[:5])

```

```python
# Create a OneHotEncoder for the Attrition column
onehot_encoder_attr = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
onehot_encoder_attr.fit(y_train[['Attrition']])


# Create two new variables by applying the encoder
# to the training and testing data
y_train_attr_encoded = onehot_encoder_attr.transform(y_train[['Attrition']])
y_test_attr_encoded = onehot_encoder_attr.transform(y_test[['Attrition']])
print(y_train_attr_encoded[:5])
print(y_test_attr_encoded[:5])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
input_dim = X_train_scaled.shape[1]
print(f"Number of columns in X training data: {input_dim}")

# Create the input layer

input_layer = Input(shape=(input_dim,))
# Create at least two shared layers
shared_layer_1 = Dense(64, activation='relu', name='shared1')(input_layer)
shared_layer_2 = Dense(128, activation='relu', name='shared2')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
dept_branch_hidden = Dense(32, activation='relu', name='department_hidden')(shared_layer_2)
dept_output = Dense(3, activation='softmax', name='department_output')(dept_branch_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
attr_branch_hidden = Dense(32, activation='relu', name='attrition_hidden')(shared_layer_2)
attr_output = Dense(2, activation='softmax', name='attrition_output')(attr_branch_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output, attr_output])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summarize the model
model.summary()
```

```python
# Train the model

history = model.fit(
    X_train_scaled,
    {'department_output': y_train_dept_encoded, 'attrition_output': y_train_attr_encoded},
    epochs=50,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(X_test_scaled,
                         {'department_output': y_test_dept_encoded, 'attrition_output': y_test_attr_encoded})

# Print the evaluation results
```

```python
# Print the accuracy for both department and attrition
print(f"Total loss: {results[0]}")
print(f"Department loss: {results[1]}")
print(f"Attrition loss: {results[2]}")
print(f"Department prediction accuracy: {results[3]}")
print(f"Attrition prediction accuracy: {results[4]}")
```

1. Is accuracy the best metric to use on this data? Why or why not?
Accuracy is a useful metric, but it might not be the best metric to use for this data. The dataset could be imbalanced with more instances of one class than the other, particularly for the attrition prediction (i.e., more employees might not leave than those who do). In such cases, accuracy can be misleading because a model could achieve high accuracy by simply predicting the majority class.
